```python
# 라이브러리 불러오기 및 api key 설정
from openai import OpenAI
from dotenv import load_dotenv
import os 

load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')

client = OpenAI(api_key=api_key)
```

```python
# 첫번째 TTS 테스트
response = client.audio.speech.create(
    model='tts-1-hd',
    voice='alloy',
    input='Hello world! This is a TTS test.',
)

response.write_to_file('hello_world.mp3') # mp3로 파일 생성

# 재생
import IPython.display as ipd

ipd.Audio('hello_world.mp3')
```

```python
# 목소리를 바꾸고 테스트
voice = 'ash'
mp3_file = f'hello_world_{voice}.mp3'

response = client.audio.speech.create(
    model='tts-1-hd',
    voice=voice,
    input=f"Hello world! I'm {voice}. This is a TTS test."
)
response.write_to_file(mp3_file)
ipd.Audio(mp3_file)
```

```python
import json

# json 파일 열기
with open('image_quiz_eng.json','r',encoding='utf-8') as f:
    eng_dict = json.load(f)

eng_dict
```

```python
voices = ['alloy','ash','coral','echo','fable','onyx','nova','sage','shimmer']

for q in eng_dict:
    no = q['no']
    quiz = q['eng']
    quiz = quiz.replace('- (1)','-One.\t')
    quiz = quiz.replace('- (2)','-Two.\t')
    quiz = quiz.replace('- (3)','-Three.\t')
    quiz = quiz.replace('- (4)','-Four.\t')
    print(no, quiz)

    # 문제 번호에 따라 목소리 선택

    voice = voices[no % len(voices)]

    response = client.audio.speech.create(
        model='tts-1-hd',
        voice=voice,
        input=f'#{no}, {quiz}',
    )
    response.write_to_file(f'audio/{no}.mp3')
```

```python
# 재생
ipd.Audio('audio/1.mp3')
```
